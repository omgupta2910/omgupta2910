```python
import random

user_action = input("Enter your choice (rock, paper, scissors): ")
possible_actions = ["rock", "paper", "scissors"]
computer_action = random.choice(possible_actions)
print("You chose {user_action}, computer chose {computer_action}.\n")

if user_action == computer_action:
    print("Both players selected {user_action}, so its a tie!")
elif user_action == "rock":
    if computer_action == "scissors":
        print("Rock sbeats scissors! You win!")
    else:
        print("Paper beats rock! You lose.")
elif user_action == "paper":
    if computer_action == "rock":
        print("Paper beats rock! You win!")
    else:
        print("Scissors beats paper! You lose.")
elif user_action == "scissors":
    if computer_action == "paper":
        print("Scissors beats paper! You win!")
    else:
        print("Rock sbeats scissors! You lose.")
```
